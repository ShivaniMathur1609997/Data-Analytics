```python
#Our data is prepared now for analysis
# Correlation -------- Relationship among different variables
# range of correlation (-1 to 1)
# Types:
# 1. Positive(+) ----- if one variable is increasing other is also increasing, decreasign also
#      a. highly correlated
#      b. nominal 
# 2. Negative(-) ------if one in increasing other is decreasing and vice versa
#      a. higly correlated
#      b. nominal
# 3. Zero ---------- no connection of one feature with other
```

```python
data.corr()
# this is known as Correlation Matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(data.corr(), cmap = "rainbow", annot = True)
plt.savefig("Correlation")
#plt.show()
```

```python
data.columns
```

```python
fig = plt.subplots(figsize = (10,15))
plt.subplot(3,2,1)
plt.subplots_adjust(hspace = 1.0)
sns.countplot(x='dept', data= data, hue = "left")
plt.xticks(rotation = 90)
plt.show()

# 0 means not leaving the company
# 1 means they are at notice period
```

```python
#Analysis from this is people from sales,technical,support at at high rate of churning
#Conclusion ----- 1. sales environment is very bad
#                 2.we need to incentivise people in sales, tech and support teams the most?
#                 3. Plese prioritize sales,technical,support
```

```python
fig = plt.subplots(figsize = (10,15))
plt.subplot(3,2,1)
plt.subplots_adjust(hspace = 1.0)
sns.countplot(x='salary', data= data, hue = "left")
plt.xticks(rotation = 90)
plt.show()
```

```python
#Analysis ---- people with low and medium salary are churning the most
#Recomendation to CEO------- provide incentives and hikes for elligible candidates
```

```python
fig = plt.subplots(figsize = (10,15))
plt.subplot(3,2,1)
plt.subplots_adjust(hspace = 1.0)
sns.countplot(x='timeSpent.company', data= data, hue = "left")
plt.xticks(rotation = 90)
plt.show()
```

```python
#Analysis: Between exp of 3 to 6 people are leaving the most
#Recommendation to CEO : Prioritize these people with their demands 
```
